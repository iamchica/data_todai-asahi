## Make a panel dataset

```python
# 政党の選択肢を揃える
# 1 自民党、2 民主党、3 維新の党、4 公明党、5 次世代の等
# 6 共産党、7 生活の党、8 社民党、9 新党改革
# 10　幸福実現党、11　支持政党なし、12　諸派、13　無所属
# 14　未来の党、15　日本維新の会、16　みんなの党、17　新党大地
# 18　国民新党、19　新党日本、20　改革クラブ、21　保守新党
# 22　無所属の会、23　自由連合、24　自由党、25　革命

data2003a['partycod2'] = data2003a['partycod'].replace({3:4, 4:24, 5:6, 6:8, 7:21, 8:22, 9:25, 10:23, 11:13})
data2003b['partycod2'] = data2003b['partycod'].replace({3:4, 4:6, 5:8, 6:21, 7:22, 8:23, 9:12, 10:13})
data2005['partycod2'] = data2005['partycod'].replace({3:4, 4:6, 5:8, 6:18, 7:19, 8:12, 9:13})
data2009['PARTY'] = data2009['PARTY'].replace({3:4, 4:6, 5:8, 6:18, 7:16, 8:19, 9:17, 10:20, 11:12, 12:13})
data2012['PARTY'] = data2012['PARTY'].replace({1:2, 2:1, 3:14, 5:15, 7:16, 9:17, 10:18, 11:19, 12:9, 13:12,14:13})
#CHECK : data2012.groupby('PARTY').count().reset_index()
print(data2003a['NAME'])
```

```python
# keep attributes like name, age, etc
data2003a_atren = data2003a.rename(columns = {
    'id':'ID','response':'RESPONSE','blockcod':'PRBLOCK','prefectu':'PREFEC','prefdist':'DISTRICT','partycod2':'PARTY'
})
data2003b_atren = data2003b.rename(columns = {
    'id':'ID','response':'RESPONSE','blockcod':'PRBLOCK','prefectu':'PREFEC','prefdist':'DISTRICT','partycod2':'PARTY',
    'incumben':'INCUMB','won':'RESULT'
})
data2005_atren = data2005.rename(columns = {
    'id':'ID','response':'RESPONSE','blockcod':'PRBLOCK','prefectu':'PREFEC','prefdist':'DISTRICT','partycod2':'PARTY',
    'sex':'SEX','age':'AGE','incumben':'INCUMB','won':'RESULT'
})

attr = ['ID','NAME','RESPONSE','PREFEC','DISTRICT','PRBLOCK','INCUMB','TERM','PARTY','SEX','AGE','RESULT']
data2003a_attr = data2003a_atren.filter(attr)
data2003b_attr = data2003b_atren.filter(attr)
data2005_attr = data2005_atren.filter(attr)
data2009_attr = data2009.filter(attr)
data2012_attr = data2012.filter(attr)
data2014_attr = data2014.filter(attr)
```

```python
# rename columns
data2003a_rename = data2003a.rename(columns = {
    'defense':'yn_defense', 'treaty':'yn_treaty','attack':'yn_preemp','un':'yn_unsc',
    'smallgov':'yn_smgov','woman':'yn_female','lifetime':'yn_lifetime','publicen':'yn_pubspend','keynes':'yn_fiscalpol'  
})
data2003b_rename = data2003b.rename(columns = {
    'defense':'yn_defense', 'treaty':'yn_treaty','attack':'yn_preemp','un':'yn_unsc','nkorea':'yn_nkorea','collecti':'yn_collecti',
    'smallgov':'yn_smgov','lifetime':'yn_lifetime','publicen':'yn_pubspend','keynes':'yn_fiscalpol','safety':'yn_safety','foreign':'yn_frgvote'
})
data2005_rename = data2005.rename(columns = {
    'defense':'yn_defense', 'treaty':'yn_treaty','attack':'yn_preemp','un':'yn_unsc','nkorea':'yn_nkorea','collecti':'yn_collecti',
    'smallgov':'yn_smgov','lifetime':'yn_lifetime','publicen':'yn_pubspend','keynes':'yn_fiscalpol','safety':'yn_safety','foreign':'yn_frgvote'
})
data2009_rename = data2009.rename(columns = {
    'Q4_1':'camp1', 'Q4_2':'camp2', 'Q4_3':'camp3',
    'Q5_1':'coal_lib', 'Q5_2':'coal_dem', 'Q5_3':'coal_kmei', 'Q5_4':'coal_com', 'Q5_5':'coal_soc', 'Q5_6':'coal_koku',
    'Q8_1':'yn_medconst', 'Q8_4':'yn_reps',
    'Q9_1':'yn_const', 'Q9_2':'yn_defense', 'Q9_4':'yn_preemp', 'Q9_5':'yn_unsc', 'Q9_6':'yn_nkorea', 'Q9_7':'yn_collecti2',
    'Q9_9':'yn_smgov', 'Q9_9':'yn_lifetime','Q9_11':'yn_pubspend', 'Q9_12':'yn_fiscalpol', 'Q9_14':'yn_tax5yrs', 'Q9_16':'yn_frgnvote', 'Q9_17':'yn_frgnwork', 'Q9_18':'yn_privacy',
    'Q10_1':'ab_env', 'Q10_7':'ab_singlemum', 'Q10_9':'ab_trade',
    'Q11_1':'fav_aso','Q11_2':'fav_yosa','Q11_3':'fav_hatokuni','Q11_4':'fav_nakagawa','Q11_5':'fav_koike',
    'Q11_6':'fav_masuzoe', 'Q11_7':'fav_hatoyuki','Q11_8':'fav_okada','Q11_9':'fav_ozawa', 'Q11_10':'fav_kan',
    'Q11_11':'fav_maehara','Q11_12':'fav_watanabe', 'Q11_13':'fav_hashimoto','Q11_14':'fav_higashi'
})
#Q2(2009)とQ1(2012)、Q3(2009)とQ3(2012)は設問は同じだが選択肢が異なる。しかも順位づけ設問なので今回は除外。
#集団自衛権については2003bと2005は文言が同じ、2009と2012はそれぞれ微妙に憲法への言及の仕方が異なる。微妙なので近い名前をつけて残す。
data2012_rename = data2012.rename(columns = {
    'Q1_1':'prim1', 'Q1_2':'prim2', 'Q1_3':'prim3',
    'Q2_1':'camp1', 'Q2_2':'camp2', 'Q2_3':'camp3',
    'Q4_1':'coal_dem', 'Q4_2':'coal_lib', 'Q4_3':'coal_ichi', 'Q4_4':'coal_kmei', 'Q4_5':'coal_com', 
    'Q4_6':'coal_min', 'Q4_7':'coal_soc', 'Q4_8':'coal_ishin', 'Q4_9':'coal_koku','Q4_10':'coal_dai', 
    'Q4_11':'coal_gen','Q4_12':'coal_sun','Q4_13':'coal_niho','Q4_14':'coal_mid','Q4_15':'coal_kai',
    'Q5_1':'yn_const', 'Q5_2':'yn_defense', 'Q5_3':'yn_preemp', 'Q5_4':'yn_unsc', 'Q5_5':'yn_nkorea', 'Q5_6':'yn_collecti3','Q5_7':'yn_smgov', 'Q5_8':'yn_pubspend', 'Q5_9':'yn_fiscalpol', 'Q5_10':'yn_tax5yrs', 'Q5_11':'yn_tax10pc', 'Q5_15':'yn_privacy', 'Q5_16':'yn_frgnvote', 'Q5_17':'yn_frgnwork', 'Q5_18':'yn_moral', 'Q5_21':'yn_reps', 'Q5_25':'yn_medconst',
    'Q6_1':'ab_asiaus', 'Q6_2':'ab_compet', 'Q6_3':'ab_trade', 'Q6_4':'ab_env', 'Q6_5':'ab_singlemum',
    'Q10_1':'fav_noda','Q10_2':'fav_abe','Q10_3':'fav_ozawa', 'Q10_4':'fav_yamaguchi','Q10_5':'fav_shii',
    'Q10_6':'fav_watanabe', 'Q10_7':'fav_fukushima','Q10_8':'fav_hashimoto', 'Q10_9':'fav_jimi','Q10_10':'fav_suzuki',
    'Q10_11':'fav_kawamura','Q10_12':'fav_ishihara','Q10_13':'fav_hiranuma','Q10_14':'fav_tanaka','Q10_15':'fav_tanioka',
    'Q10_16':'fav_masuzoe','Q10_17':'fav_koshiishi','Q10_18':'fav_ishiba',
    'Q11_1':'media_hp', 'Q11_2':'media_bul','Q11_3':'media_sns', 'Q11_4':'media_tv'
})
data2014_rename = data2014.rename(columns = {
    'Q1_1':'prim1', 'Q1_2':'prim2', 'Q1_3':'prim3',
    'Q2_1':'camp1', 'Q2_2':'camp2', 'Q2_3':'camp3',
    'Q6_1':'yn_defense', 'Q6_2':'yn_preemp', 'Q6_3':'yn_nkorea', 'Q6_5':'yn_smgov', 'Q6_6':'yn_pubspend', 'Q6_7':'yn_fiscalpol', 'Q6_8':'yn_tax10pc', 'Q6_11':'yn_privacy', 'Q6_12':'yn_frgnvote', 'Q6_13':'yn_frgnwork', 'Q6_14':'yn_moral','Q6_17':'yn_female',
    'Q7_1':'ab_asiaus', 'Q7_2':'ab_compet', 'Q7_3':'ab_trade', 'Q7_4':'ab_env', 'Q7_5':'ab_singlemum',
    'Q14_1':'media_hp', 'Q14_2':'media_sns', 'Q14_3':'media_tv'
})
# keep columns (consistent survey questions)
data2003a_subset = data2003a_rename.loc[:,data2003a_rename.columns.str.contains('yn_|media_|ideology')]
data2003b_subset = data2003b_rename.loc[:,data2003b_rename.columns.str.contains('yn_|media_|ideology|constitu')]
data2005_subset = data2005_rename.loc[:,data2005_rename.columns.str.contains('yn_|media_|ideology|constitu')]
data2009_subset = data2009_rename.loc[:,data2009_rename.columns.str.contains('camp|coal_|yn_|ab_|fav_')]
data2012_subset = data2012_rename.loc[:,data2012_rename.columns.str.contains('prim|camp|coal_|yn_|ab_|fav_|media_')]
data2014_subset = data2014_rename.loc[:,data2014_rename.columns.str.contains('prim|camp|yn_|ab_|media_')]
```

```python
# append datasets by row 
data2003a_use = pd.concat([data2003a_attr,data2003a_subset],axis=1)
data2003b_use = pd.concat([data2003b_attr,data2003b_subset],axis=1)
data2005_use = pd.concat([data2005_attr,data2005_subset],axis=1)
data2009_use = pd.concat([data2009_attr,data2009_subset],axis=1)
data2012_use = pd.concat([data2012_attr,data2012_subset],axis=1)
data2014_use = pd.concat([data2014_attr,data2014_subset],axis=1)
```

```python
# add election year
data2003a_use['YEAR'] = '2003'
data2003b_use['YEAR'] = '2003'
data2005_use['YEAR'] = '2005'
data2009_use['YEAR'] = '2009'
data2012_use['YEAR'] = '2012'
data2014_use['YEAR'] = '2014'

data2003b_use['ELECYEAR'] = '2003'
data2005_use['ELECYEAR'] = '2005'
data2009_use['ELECYEAR'] = '2009'
data2012_use['ELECYEAR'] = '2012'
data2014_use['ELECYEAR'] = '2014'
```

```python
data = pd.concat([data2003a_use, data2003b_use, data2005_use, data2009_use, data2012_use, data2014_use])
# remove irrelevant spaces from names
data['NAME'] = data['NAME'].astype(str).apply(lambda x: x.strip()) 

# concat names and prefecture numbers to make an id
data['cat'] = data['NAME'].astype('category') # convert
data['cat'] = data['cat'].cat.codes
data['uid'] = data[['cat','PREFEC']].astype(str).apply(lambda x: '.'.join(x), axis=1) # concatenate names and prefecture numbers. Try '='.join(x).
del(data['cat'])
```

```python
data = data[[
    'uid',
    'ID',
    'YEAR',
    'ELECYEAR',
    'NAME',
    'RESPONSE',
    'PREFEC',
    'DISTRICT',
    'PRBLOCK',
    'INCUMB',
    'TERM',
    'PARTY',
    'SEX',
    'AGE',
    'RESULT',
    'prim1',
    'prim2',
    'prim3',
    'camp1',
    'camp2',
    'camp3',
    'coal_lib',
    'coal_dem',
    'coal_kmei',
    'coal_com',
    'coal_soc',
    'coal_koku',
    'constitu',
    'ideology',
    'yn_medconst',
    'yn_reps',
    'yn_const',
    'yn_defense',
    'yn_preemp',
    'yn_unsc',
    'yn_nkorea',
    'yn_smgov',
    'yn_pubspend',
    'yn_fiscalpol',
    'yn_female',
    'yn_tax5yrs',
    'yn_frgnvote',
    'yn_frgnwork',
    'yn_privacy',
    'yn_tax10pc',
    'yn_moral',
    'media_hp',
    'media_sns',
    'media_bul',
    'media_tv',
    'ab_asiaus',
    'ab_compet',
    'ab_env',
    'ab_singlemum',
    'ab_trade',
    'fav_masuzoe',
    'fav_ozawa',
    'fav_watanabe',
    'fav_hashimoto',
]]
```
